```python
with open("data/transcript.txt", "r") as f:
    transcript = f.read()

print(transcript[:500])
```

```python
def split_speakers(text):
    doctor = []
    patient = []

    for line in text.split("\n"):
        line = line.strip()
        if line.startswith("Physician:"):
            doctor.append(line.replace("Physician:", "").strip())
        elif line.startswith("Patient:"):
            patient.append(line.replace("Patient:", "").strip())

    return " ".join(doctor), " ".join(patient)

doctor_text, patient_text = split_speakers(transcript)

print("Patient text preview:\n", patient_text[:300])
```

```python
MEDICAL_RULES = {
    "Symptoms": ["pain", "discomfort", "backache", "neck", "stiffness"],
    "Diagnosis": ["whiplash"],
    "Treatment": ["physiotherapy", "painkillers", "x-ray"],
    "Prognosis": ["recovery", "six months", "no long-term"]
}

def extract_medical_entities(text):
    extracted = {key: [] for key in MEDICAL_RULES}

    sentences = text.lower().split(".")
    for sentence in sentences:
        for category, keywords in MEDICAL_RULES.items():
            if any(word in sentence for word in keywords):
                extracted[category].append(sentence.strip())

    return extracted

entities = extract_medical_entities(transcript)
entities
```

```python
def detect_sentiment(text):
    text = text.lower()
    if any(word in text for word in ["worried", "concern", "anxious"]):
        return "Anxious"
    elif any(word in text for word in ["relief", "great", "good", "reassuring"]):
        return "Reassured"
    else:
        return "Neutral"

sentiment = detect_sentiment(patient_text)
sentiment
```

```python
def detect_intent(text):
    text = text.lower()
    if "worried" in text or "future" in text:
        return "Seeking reassurance"
    elif "pain" in text or "hurt" in text:
        return "Reporting symptoms"
    else:
        return "General follow-up"

intent = detect_intent(patient_text)
intent
```
